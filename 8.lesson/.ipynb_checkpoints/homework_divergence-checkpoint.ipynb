## Логистическая регрессия

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import log_loss

%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,5)
```

```python
np.random.seed(0)

C1 = np.array([[0., -0.8], [1.5, 0.8]])
C2 = np.array([[1., -0.7], [2., 0.7]])
gauss1 = np.dot(np.random.randn(200, 2) + np.array([5, 3]), C1)
gauss2 = np.dot(np.random.randn(200, 2) + np.array([1.5, 0]), C2)

X = np.vstack([gauss1, gauss2])
y = np.r_[np.ones(200), np.zeros(200)]
print(X[:5])
print(y[:5])
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
from sklearn.base import BaseEstimator, ClassifierMixin


class MySGDClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, alpha=0.01, max_epoch=10):
        self.alpha = alpha
        self.max_epoch = max_epoch
        self.loses = []
        
    def sigmoid(self, scores):
        return 1 / (1 + np.exp(-scores))          
    
    def logloss(self, X, y):
        yt = np.dot(X, self.theta)
        p = self.sigmoid(yt)
        eps = 1e-15
        p = np.clip(p, eps, 1 - eps) 
        loss = - np.sum(y * np.log(p) + (1 - y) * np.log(1 - p)) / len(X)
        return loss
    
    def logloss2(self, X, y):
        yt = np.dot(X, self.theta)
        p = self.sigmoid(yt)
        return log_loss(y, p)
    
    def fit(self,X, y=None):
        X_b = np.c_[np.ones((len(X), 1)), X]
        m = len(X)
        alpha = self.alpha
        n_epochs = self.max_epoch
        eps = 0.000001

        self.theta = np.random.randn(3, 1)  # random initialization
        it = 0
        for epoch in range(n_epochs):
            p = np.random.permutation(m)
            for idx in p:
                random_index = np.random.randint(m)
                xi = X_b[[idx], :]
                yi = y[idx]
                gradients = 2 * xi.T.dot(self.sigmoid(xi.dot(self.theta)) - yi)
                theta_old  = self.theta
                self.theta = self.theta - alpha * gradients
                if it % 100 == 0:
                    self.loses.append(self.logloss(X_b, y))
                it += 1
        return 
        
    def predict(self, X):
        return np.sign(np.c_[X, np.ones((len(X), 1))].dot(self.theta))
```

```python
#X[[11],:]
#X
model = MySGDClassifier(alpha=0.00005, max_epoch=100)
model.fit(X, y)
model.theta
```

```python
x_naive = np.array([-4, 4])
# Ax + By + C -> y = (-AX -C) / B
y_naive = (-model.theta[1] * x_naive - model.theta[0] ) / model.theta[2]
plt.scatter(X[:,0], X[:,1], c=y)
plt.plot(x_naive, y_naive, c='red', label='descision boundary naive')
```

```python
%matplotlib inline
import numpy as np

import matplotlib.pyplot as plt
import random
from matplotlib import cm
from sklearn.metrics import log_loss

eps = 1e-15

def sigmoid(scores):
    return 1. / (1 + np.exp(-scores))  

def fun(theta, X, y):
    yt = np.dot(X, theta)
    p = sigmoid(yt)
    eps = 1e-15
    p = np.clip(p, eps, 1 - eps) 
    loss = - np.mean(y * np.log(p) + (1 - y) * np.log(1 - p))
    return loss

def fun_loss(theta, X, y):
    yt = np.dot(X, theta)
    p = sigmoid(yt)
    return log_loss(y, p)

def plot_cnt(X_train, y_train):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    x = y = np.arange(-20.0, 20.0, 1.)
    X, Y = np.meshgrid(x, y)
    zs = np.array([fun(np.array([x,y]), X_train, y_train) for x,y in zip(np.ravel(X), np.ravel(Y))])
    Z = zs.reshape(X.shape)
    plt.contourf(X,Y,Z) #reshape Z too!
    plt.colorbar()

plot_cnt(X, y)
```
